```python
import pandas as pd
import numpy as np

# Generate synthetic code snippets
num_samples = 250
code_snippets = []
for _ in range(num_samples):
    # Generate random code snippet (you can replace this with actual code examples)
    snippet_length = np.random.randint(5, 20)
    code = ''.join(np.random.choice(list('abcdefghijklmnopqrstuvwxyz '), size=snippet_length))
    code_snippets.append(code)

# Generate corresponding quality labels
quality_labels = np.random.choice(['bug', 'style violation', 'performance issue'], size=num_samples)
print("Code_snippets",code_snippets )
print("quality_labels",quality_labels)

# Create a DataFrame
```

```python
df = pd.DataFrame({'code_snippets': code_snippets, 'quality_label': quality_labels})

# Save the dataset to a CSV file
df.to_csv('code_review_dataset.csv', index=False)


e= pd.read_csv('code_review_dataset.csv')
e.head(50)
```
